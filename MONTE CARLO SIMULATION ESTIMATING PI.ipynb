## ESTIMATING PI

### PI IS THE RATIO B/W AREA AND RADIUS SQUARE

```python
import random as r
from math import sqrt
```

```python
def distance(a,b):
    return sqrt((a**2) + (b**2) )
def estimate_pi(darts):
    inside=0
    
    for i in range(darts):
        x=r.random()
        y=r.random()
        if distance(x,y)<1.0:
            inside+=1
            
    pi=(inside/darts)*4
    return pi
```

```python
num_sim=int(1e+7)
pi=estimate_pi(num_sim)
print(pi)
```

# RANDOM WALKS

```python
def get_action():
    return r.choice(  [(0,1),(0,-1),(1,0),(-1,0)]  )
def random_walk(n):
    path=[]
    x,y=0,0
    for i in range(n):
        (dx,dy)=get_action()
        x=x+dx
        y=y+dy
        
        path.append( (x,y) )
    return path
    
```

```python
steps=2000
walk=random_walk(steps)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(num=None,figsize=(15,10))
point_color=list(range(steps))
x_numbers=[x[0] for x in walk]
y_numbers=[x[1] for x in walk]
plt.scatter(x_numbers,y_numbers,marker='.',s=100,c=point_color)
plt.show()
```
